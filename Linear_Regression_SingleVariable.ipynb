<h2>Linear Regression Using Single Variable</h2>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('gdp.csv')
df
```

```python
%matplotlib inline
plt.ylabel('GDP(US$)')
plt.xlabel('Year')
plt.title('Canada per capita income')
plt.scatter(df.year, df['per capita income (US$)'])
```

```python
# training a model using sklearn
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
d = pd.read_csv('years.csv')
d
```

```python
# testing data for next years
gdp = reg.predict(d)
d['gdp'] = gdp
d
```

```python
# slope of line
reg.coef_
```

```python
# y-intercept of line
reg.intercept_
```

```python
# line 
%matplotlib inline
plt.xlabel('GDP(US$)')
plt.ylabel('Year')
plt.title('Canada per capita income')
plt.scatter(df.year,df['per capita income (US$)'])
plt.plot(df.year,reg.predict(df[['year']]))
```
