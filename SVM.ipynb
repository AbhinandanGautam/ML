#Homeprices Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.costly=='yes']
df1 = df[df.costly=='no']
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df0['area'],df0['price'],color='green',marker='+')
plt.scatter(df1['area'],df1['price'],color='blue',marker='*')
```

```python
X = df[['area', 'price']]
y = df[['costly']]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

#CM1 Dataset

```python
ndf = pd.read_csv('cm1.csv')
ndf.head()
```

```python
X = ndf.drop('defects', axis=1)
X.head()
```

```python
y = ndf[['defects']]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

#Digits Dataset

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
digits.target
```

```python
dir(digits)
```

```python
digits.target_names
```

```python
df = pd.DataFrame(digits.data,digits.target)
df.head()
```

```python
df['target'] = digits.target
df.head(20)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target',axis='columns'), df.target, test_size=0.3)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
